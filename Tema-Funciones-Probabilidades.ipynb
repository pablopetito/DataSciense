##### La simulacion de Monte Carlo

Probabilidad de caer dentro del circulo 
Teniendo un cuadrado de 400x400 y dentro un circulo con
radio 200
Que probabilidad hay de elegir un punto y que caiga dentro
del circulo?
P(probabilidad)=CaerDentrodelCirculo / caerDentrodelCuadrado
P = area circulo / area cuadrado
P = Pi . r . r / 2.r . 2.r # se simplifica
P = Pi / 4

```python
#generamos dos numeros aleatorios x , y entre 0 y 1
#calcularemos x  * x + y * y 
    # Si el valor es inferior a 1 -> dentro del circulo
    # Si el valor es superior a 1 -> fuera del circulo
#calculamos el numero total de veces que estan dentro del
#y lo dividimos entreel numero total de intentos para 
#obtener una aproximacion de la probabilidad dentro del ciculo
#Usamos dicha probabilidad para aproximar el valor de pi
#Repetimos el experimento un numero suficiente de veces
#para obtener diferentes aproximaciones de pi
#calcular el promedio del experimento para dar un valor final
```

```python
def pi_montecarlo(n, n_exp):
    pi_avg = 0
    pi_value_list = []
    for i in range(n_exp):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt(x[j] * x[j] + y[j] * y[j])
            if z <=1:
                value +=1
        float_value = float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value

    pi = pi_avg/n_exp

    print(pi)
    fig = plt.plot(pi_value_list)
    return (pi, fig)
    
        
```

```python
pi_montecarlo(1000, 200)
```
