```python
def randint_list(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint_list(25, 1, 50)
```

```python
import random 
for i in range(10):
    print(random.randrange(0,100,3))
```
